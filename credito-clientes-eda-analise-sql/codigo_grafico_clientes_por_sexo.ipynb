```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dados do CSV em um DataFrame
df = pd.read_csv('clientes_por_sexo.csv')

# Calculando as porcentagens
total = df['_col0'].sum()
percentagens = [(count / total) * 100 for count in df['_col0']]

# Configurando as labels
labels = ['{} - {:.1f}%'.format(sexo, percentual) for sexo, percentual in zip(df['sexo'], percentagens)]

# Criando o gráfico de pizza
plt.pie(df['_col0'], labels=labels, autopct='%1.1f%%', startangle=140)

# Adicionando título
plt.title('Distribuição de Sexo')

# Mostrando o gráfico
plt.axis('equal')  # Garante que o gráfico seja desenhado como um círculo
plt.show()
```
